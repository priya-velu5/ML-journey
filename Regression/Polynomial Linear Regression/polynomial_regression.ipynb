# Polynomial Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

## Training the Polynomial Regression model on the whole dataset

How to do polynomial regression on this? 
polynomial linear regression eqn: y = b0 + b1x + b2x^2 + b3x^3 ...... bnx^n 

So in order to do polynomial regression, create the matrix with all the required powers. For eg: power = 3

[ Xi Xi^2 Xi^3] -  denotes one row of the independent variable(s). 
Now, we can apply the same linear regressor model on it. 

```python
from sklearn.preprocessing  import PolynomialFeatures 
# make the polynomial input 
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)

# make the model now
reg2 = LinearRegression()
reg2.fit(X_poly,y)
```

## Visualising the Linear Regression results

```python
#scatter plot the points. 
#plot the regression line
y_pred = lin_reg.predict(X)

plt.scatter(X,y,color='red')
plt.plot(X,y_pred)
plt.title("Linear Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results

```python
#scatter plot the points. 
#plot the regression line
yp_pred = reg2.predict(X_poly)

plt.scatter(X,y,color='red')
plt.plot(X,yp_pred)
plt.title("Polynomial Linear Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```python
#retraining the model for higher degrees of X
from sklearn.preprocessing  import PolynomialFeatures 
# make the polynomial input 
poly_reg2 = PolynomialFeatures(degree = 5)
X_poly2 = poly_reg2.fit_transform(X)

# make the model now
reg3 = LinearRegression()
reg3.fit(X_poly2,y)
```

```python
#scatter plot the points. 
#plot the regression line
yp_pred2 = reg3.predict(X_poly2)

plt.scatter(X,y,color='red')
plt.plot(X,yp_pred2)
plt.title("Polynomial Linear Regression")
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
# for a finer plot : make the scale finer. 
# All the numbers from least to the highest observation of X increamenting at 0.1, are considered to plot the line. 
X_grid = np.arange(min(X),max(X),0.1)
X_grid = X_grid.reshape((len(X_grid),1))
```

```python
#Making a single prediction (for example the salary of an employee with position 6.5)
print(lin_reg.predict([[6.5]]))
```

```python
#Making a single prediction (for example the salary of an employee with position 6.5)
print(reg2.predict( poly_reg.fit_transform([[6.5]])))
```
